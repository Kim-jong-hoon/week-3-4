```python
# 3.
import cv2
import yt_dlp
import matplotlib.pyplot as plt
import tempfile
import os
import time
from IPython.display import clear_output

def play_youtube_video(youtube_url, skip_frames=1):
    """
    YouTube ÏòÅÏÉÅÏùÑ Îã§Ïö¥Î°úÎìúÌïòÍ≥† Ïû¨ÏÉùÌïòÎäî Í∞ÑÎã®Ìïú Ìï®Ïàò
    skip_frames: ÌîÑÎ†àÏûÑ Í±¥ÎÑàÎõ∞Í∏∞ (1=Î™®Îì† ÌîÑÎ†àÏûÑ, 2=Ìïú ÌîÑÎ†àÏûÑÏî© Í±¥ÎÑàÎõ∞Í∏∞)
    """

    ydl_opts = {
        'format': 'mp4/best[height<=480]',  # 480pÎ°ú Ï†úÌïú (Îπ†Î•∏ Ï≤òÎ¶¨)
        'outtmpl': tempfile.gettempdir() + '/temp_video.%(ext)s',
        'quiet': True,
    }

    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            print("‚¨áÔ∏è ÏòÅÏÉÅ Îã§Ïö¥Î°úÎìú Ï§ë...")
            info = ydl.extract_info(youtube_url, download=True)

            video_path = ydl.prepare_filename(info)

            cap = cv2.VideoCapture(video_path)
            fps = cap.get(cv2.CAP_PROP_FPS)
            frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

            print(f"üé¨ '{info['title']}' Ïû¨ÏÉù ÏãúÏûë!")
            print("(Ctrl+CÎ°ú Ï§ëÎã®)")

            frame_num = 0
            while True:
                ret, frame = cap.read()
                if not ret:
                    break

                # ÌîÑÎ†àÏûÑ Í±¥ÎÑàÎõ∞Í∏∞
                if frame_num % skip_frames == 0:
                    clear_output(wait=True)
                    plt.figure(figsize=(10, 6))
                    plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
                    plt.title(f"Ïû¨ÏÉù Ï§ë... ({frame_num}/{frame_count})")
                    plt.axis('off')
                    plt.show()

                    # Ïû¨ÏÉù ÏÜçÎèÑ Ï°∞Ï†à
                    time.sleep(0.1)  # 0.1Ï¥à ÎåÄÍ∏∞ (Îπ†Î•∏ Ïû¨ÏÉù)

                frame_num += 1

            cap.release()
            os.remove(video_path)
            print("‚úÖ Ïû¨ÏÉù ÏôÑÎ£å!")

    except KeyboardInterrupt:
        print("\n‚èπÔ∏è Ïû¨ÏÉù Ï§ëÎã®")
    except Exception as e:
        print(f"‚ùå Ïò§Î•ò: {e}")

# ÏÇ¨Ïö©Î≤ï
youtube_url = input("YouTube URL: ")
play_youtube_video(youtube_url, skip_frames=2)  # Ìïú ÌîÑÎ†àÏûÑÏî© Í±¥ÎÑàÎõ∞Ïñ¥ Îπ†Î•∏ Ïû¨ÏÉù
```

```python
# 4.
import cv2
from google.colab import files
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import clear_output
import time

# openCV Ìï®ÏàòÎ•º Ïù¥Ïö©Ìïú ÏòÅÏÉÅ ÏóÖÎ°úÎìú
#uploaded = files.upload()
video_path = "/content/Í∞ïÎÇ®ÎåÄÎ°ú Í∞ïÎÇ®Ïó≠ ÎèÑÎ°ú ÎìúÎùºÏù¥Î∏å ÏµúÍ∑º Î∞§Í±∞Î¶¨ Î™®Ïäµ ÏûÖÎãàÎã§. [tEtWnGwwCEc].mp4"

# ÏòÅÏÉÅ Ïó¥Í∏∞
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))  # ÏÉàÎ°ú Ï†ïÏùòÌïú widthÏô∏ heightÎì§
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

print(f"üìπ ÏòÅÏÉÅ Ï†ïÎ≥¥: {frame_count}ÌîÑÎ†àÏûÑ, {fps:.2f}FPS")
print("üé¨ Ï∞®ÏÑ† Ïù∏Ïãù ÏãúÏûë! (Ctrl+CÎ°ú Ï§ëÎã®)")

frame_num = 0

# Í≤∞Í≥º ÏòÅÏÉÅÏùÑ Ï†ÄÏû•ÌïòÍ∏∞ÏúÑÌï¥ gptÎ°úÎ∂ÄÌÑ∞ Ï∂îÍ∞ÄÌïú ÏΩîÎìú, mp4ÌååÏùºÎ°ú Ï†ÄÏû•ÌïòÍ∏∞ÏúÑÌï¥ ÏÇ¨Ïö©ÌïòÎäî Î¨∏Î≤ïÏù¥Î©∞, width, heightÎèÑ Îî∞Î°ú Ï†ïÏùòÌï¥Ï§òÏïºÌïúÎã§.
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('/content/lane_detection_result.mp4', fourcc, fps, (width, height))

try:
    while True:
        ret, frame = cap.read()  #Ïπ¥Î©îÎùº (ÌòπÏùÄ ÎπÑÎîîÏò§ ÌååÏùº)Î°úÎ∂ÄÌÑ∞ ÌîÑÎ†àÏûÑÏùÑ ÏùΩÏñ¥ÏôÄ frame Ïóê Ï†ÄÏû•
        if not ret:
            print("üì∫ ÏòÅÏÉÅ Ïû¨ÏÉù ÏôÑÎ£å")
            break

        # 1. Í∑∏Î†àÏù¥Ïä§ÏºÄÏùº Î≥ÄÌôò
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # 2. Î∏îÎü¨ ‚Üí ÏóêÏßÄ(Canny) - ÏûÑÍ≥ÑÍ∞í ÎÇÆÏ∂§
        blur = cv2.GaussianBlur(gray, (5, 5), 0)
        edges = cv2.Canny(blur, 30, 100)  # 50,150 ‚Üí 30,100

        # 3. ROI ÏÑ§Ï†ï - ÌïòÎã® 30%
        height, width = edges.shape
        mask = np.zeros_like(edges)
        polygon = np.array([[
            (int(width * 0.1), height),            # ÏôºÏ™Ω ÏïÑÎûò
            (int(width * 0.9), height),            # Ïò§Î•∏Ï™Ω ÏïÑÎûò
            (int(width * 0.6), int(height * 0.7)), # Ïò§Î•∏Ï™Ω ÏúÑ (70% ÏßÄÏ†ê)
            (int(width * 0.4), int(height * 0.7))  # ÏôºÏ™Ω ÏúÑ (70% ÏßÄÏ†ê)
        ]])
        cv2.fillPoly(mask, polygon, 255)
        roi = cv2.bitwise_and(edges, mask)

        # 4. Hough TransformÏúºÎ°ú ÏßÅÏÑ† Í≤ÄÏ∂ú - ÌååÎùºÎØ∏ÌÑ∞ ÏôÑÌôî
        lines = cv2.HoughLinesP(roi, 2, np.pi / 180,
                               threshold=30,      # 50 ‚Üí 30
                               minLineLength=30,  # 40 ‚Üí 30
                               maxLineGap=80)     # 50 ‚Üí 80

        # 5. ÏõêÎ≥∏ frameÏùÑ Î≥µÏÇ¨Ìï¥ÏôÄ, Ï∞®ÏÑ† Í∑∏Î¶¨Í∏∞
        line_image = frame.copy()
        if lines is not None:
            for line in lines:
                x1, y1, x2, y2 = line[0]
                cv2.line(line_image, (x1, y1), (x2, y2), (0, 255, 0), 5)

        out.write(line_image)  # line_imageÎùºÎäî Î≥ÄÏàòÎäî Ï∞®ÏÑ†Ïù¥ Í∑∏Î†§ÏßÑ ÌîÑÎ†àÏûÑ! Ïù¥Î•º Í≤∞Í≥ºÏòÅÏÉÅÏóê Ï∂îÍ∞ÄÌïúÎã§.

        # 6. Í≤∞Í≥º Ï∂úÎ†•
        clear_output(wait=True)
        plt.figure(figsize=(12, 6))
        plt.subplot(1, 2, 1)
        plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        plt.title(f"ÏõêÎ≥∏ ÌîÑÎ†àÏûÑ {frame_num}")
        plt.axis('off')

        plt.subplot(1, 2, 2)
        plt.imshow(cv2.cvtColor(line_image, cv2.COLOR_BGR2RGB))
        plt.title(f"Ï∞®ÏÑ† Ïù∏Ïãù Í≤∞Í≥º ({len(lines) if lines is not None else 0}Í∞ú)")
        plt.axis('off')
        plt.show()

        frame_num += 1
        time.sleep(10)  # 10Ï¥àÎßàÎã§ Ïù¥Í±∏ Î∞îÍøîÎ≥∏Îã§.

except KeyboardInterrupt:
    print("\n‚èπÔ∏è Ïû¨ÏÉù Ï§ëÎã®")

cap.release()
out.release()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import time
from IPython.display import clear_output
from google.colab import files

# ÏòÅÏÉÅ Í≤ΩÎ°ú
video_path = "/content/Í∞ïÎÇ®ÎåÄÎ°ú Í∞ïÎÇ®Ïó≠ ÎèÑÎ°ú ÎìúÎùºÏù¥Î∏å ÏµúÍ∑º Î∞§Í±∞Î¶¨ Î™®Ïäµ ÏûÖÎãàÎã§. [tEtWnGwwCEc].mp4"
cap = cv2.VideoCapture(video_path)

# ÏòÅÏÉÅ Ï†ïÎ≥¥ Ï∂îÏ∂ú
fps = cap.get(cv2.CAP_PROP_FPS)
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

print(f"üìπ ÏòÅÏÉÅ Ï†ïÎ≥¥: {frame_count}ÌîÑÎ†àÏûÑ, {fps:.2f}FPS")
print("üé¨ Ï∞®ÏÑ† Ïù∏Ïãù ÏãúÏûë! (Ctrl+CÎ°ú Ï§ëÎã®)")

# Í≤∞Í≥º ÏòÅÏÉÅ Ï†ÄÏû• ÏÑ§Ï†ï
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
output_path = "/content/lane_detection_result.mp4"
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

frame_num = 0

try:
    while True:
        ret, frame = cap.read()
        if not ret:
            print("üì∫ ÏòÅÏÉÅ Ïû¨ÏÉù ÏôÑÎ£å")
            break

        # Ï∞®ÏÑ† Í≤ÄÏ∂ú
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        blur = cv2.GaussianBlur(gray, (5, 5), 0)
        edges = cv2.Canny(blur, 30, 100)

        mask = np.zeros_like(edges)
        polygon = np.array([[
            (int(width * 0.1), height),
            (int(width * 0.9), height),
            (int(width * 0.6), int(height * 0.7)),
            (int(width * 0.4), int(height * 0.7))
        ]])
        cv2.fillPoly(mask, polygon, 255)
        roi = cv2.bitwise_and(edges, mask)

        lines = cv2.HoughLinesP(roi, 2, np.pi / 180, 30, minLineLength=30, maxLineGap=80)

        line_image = frame.copy()
        if lines is not None:
            for line in lines:
                x1, y1, x2, y2 = line[0]
                cv2.line(line_image, (x1, y1), (x2, y2), (0, 255, 0), 5)

        # ÌîÑÎ†àÏûÑ Ï†ÄÏû•
        out.write(line_image)

        # Ï∂úÎ†•
        clear_output(wait=True)
        progress = (frame_num + 1) / frame_count * 100
        print(f"üìä ÏßÑÌñâÎ•†: {frame_num + 1}/{frame_count} ({progress:.1f}%)")
        print("‚èπÔ∏è Ïû¨ÏÉù Ï§ëÎã® - Í≤∞Í≥º Ï†ÄÏû• Ï§ë..." if cv2.waitKey(1) & 0xFF == ord('q') else "")

        plt.figure(figsize=(12, 6))
        plt.subplot(1, 2, 1)
        plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        plt.title(f"ÏõêÎ≥∏ {frame_num}")
        plt.axis('off')

        plt.subplot(1, 2, 2)
        plt.imshow(cv2.cvtColor(line_image, cv2.COLOR_BGR2RGB))
        plt.title(f"Ï∞®ÏÑ† Ïù∏Ïãù ({len(lines) if lines is not None else 0}Í∞ú)")
        plt.axis('off')
        plt.show()

        frame_num += 1
        time.sleep(0.05)  # ÏÜçÎèÑ Ï°∞Ï†à

except KeyboardInterrupt:
    print("‚èπÔ∏è Ïû¨ÏÉù Ï§ëÎã® - Í≤∞Í≥º Ï†ÄÏû• Ï§ë...")

# Î¶¨ÏÜåÏä§ Ï†ïÎ¶¨
cap.release()
out.release()

# Ï†ÄÏû• Î∞è Îã§Ïö¥Î°úÎìú Î©îÏãúÏßÄ
print("‚úÖ Ï≤òÎ¶¨ ÏôÑÎ£å!")
print(f"üìÅ Í≤∞Í≥º ÏòÅÏÉÅÏù¥ Ï†ÄÏû•ÎêòÏóàÏäµÎãàÎã§: {output_path}")
print("üíæ Í≤∞Í≥º ÏòÅÏÉÅ Îã§Ïö¥Î°úÎìú Ï§ë...")

# Îã§Ïö¥Î°úÎìú
files.download(output_path)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# ÏòÅÏÉÅ Ìïú ÌîÑÎ†àÏûÑ Î∂àÎü¨Ïò§Í∏∞
# Use the correct video path from the previous cell
video_path = "/content/Í∞ïÎÇ®ÎåÄÎ°ú Í∞ïÎÇ®Ïó≠ ÎèÑÎ°ú ÎìúÎùºÏù¥Î∏å ÏµúÍ∑º Î∞§Í±∞Î¶¨ Î™®Ïäµ ÏûÖÎãàÎã§. [tEtWnGwwCEc].mp4"
cap = cv2.VideoCapture(video_path)

# Check if the video opened successfully
if not cap.isOpened():
    print(f"Error: Could not open video file {video_path}")
    cap.release()
else:
    ret, frame = cap.read()
    cap.release()

    if not ret:
        print("Error: Could not read the first frame from the video.")
    else:
        # Í∏∞Î≥∏ Ï†ïÎ≥¥
        height, width = frame.shape[:2]
        original = frame.copy()

        # 1. ROI ÏòÅÏó≠
        roi_frame = frame.copy()
        polygon = np.array([[(int(width*0.1), height), (int(width*0.9), height),
                             (int(width*0.6), int(height*0.7)), (int(width*0.4), int(height*0.7))]])
        cv2.polylines(roi_frame, polygon, isClosed=True, color=(255, 0, 0), thickness=3)

        # 2. Í∑∏Î†àÏù¥ + Ïó£ÏßÄ
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        blur = cv2.GaussianBlur(gray, (5, 5), 0)
        edges = cv2.Canny(blur, 30, 100)

        # 3. ROI ÎßàÏä§ÌÅ¨
        mask = np.zeros_like(edges)
        cv2.fillPoly(mask, polygon, 255)

        # 4. ROI ÎÇ¥Î∂Ä Ïó£ÏßÄ
        roi_edge = cv2.bitwise_and(edges, mask)

        # 5. Ï†ÑÏ≤¥ ÏÑ† Í≤ÄÏ∂ú
        total_line = np.zeros_like(frame)
        lines = cv2.HoughLinesP(edges, 2, np.pi/180, 30, minLineLength=30, maxLineGap=80)
        if lines is not None:
            for x1, y1, x2, y2 in lines[:, 0]:
                cv2.line(total_line, (x1, y1), (x2, y2), (0, 255, 0), 2)

        # 6. ROI ÎÇ¥Î∂Ä ÏÑ†Îßå ÌïÑÌÑ∞ÎßÅ
        filtered_line = np.zeros_like(frame)
        if lines is not None:
            for x1, y1, x2, y2 in lines[:, 0]:
                # Check if both points are within the mask
                if mask[y1, x1] > 0 and mask[y2, x2] > 0:
                    cv2.line(filtered_line, (x1, y1), (x2, y2), (0, 255, 0), 2)


        # 7. ÏµúÏ¢Ö Í≤∞Í≥º
        final = frame.copy()
        if lines is not None:
            for x1, y1, x2, y2 in lines[:, 0]:
                 # Check if both points are within the mask
                if mask[y1, x1] > 0 and mask[y2, x2] > 0:
                    cv2.line(final, (x1, y1), (x2, y2), (0, 255, 0), 3)

        # --------------------------------------------
        # üìä Ï†ÑÏ≤¥ subplot ÏãúÍ∞ÅÌôî
        fig, axs = plt.subplots(2, 4, figsize=(20, 8))

        axs[0, 0].imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
        axs[0, 0].set_title('original frame 0')
        axs[0, 1].imshow(cv2.cvtColor(roi_frame, cv2.COLOR_BGR2RGB))
        axs[0, 1].set_title('ROI boundary blue')
        axs[0, 2].imshow(edges, cmap='gray')
        axs[0, 2].set_title('total edge detection')
        axs[0, 3].imshow(mask, cmap='gray')
        axs[0, 3].set_title('ROI mask')

        axs[1, 0].imshow(roi_edge, cmap='gray')
        axs[1, 0].set_title('ROI after edge')
        axs[1, 1].imshow(cv2.cvtColor(total_line, cv2.COLOR_BGR2RGB))
        axs[1, 1].set_title('detection total line')
        axs[1, 2].imshow(cv2.cvtColor(filtered_line, cv2.COLOR_BGR2RGB))
        axs[1, 2].set_title('filtering lane')
        axs[1, 3].imshow(cv2.cvtColor(final, cv2.COLOR_BGR2RGB))
        axs[1, 3].set_title('final result (50ÌîÑÎ†àÏûÑ ÏòàÏãú)')

        for ax in axs.ravel():
            ax.axis('off')

        plt.tight_layout()
        plt.show()
```
